```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Training_data_uhi_index_2025-02-18.csv")
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.describe()
```

```python
new_df=df.dropna()

print(new_df.to_string())
```

```python
print(new_df.duplicated())
```

```python
new_df = new_df.drop_duplicates()

print(new_df.to_string())
```

```python
print(new_df.dtypes)
```

```python
sns.scatterplot(x="datetime", y="UHI Index", data=new_df)

plt.title("Grfico de disperso entre data e indice de UHI")
plt.xlabel("Data e Hora")
plt.ylabel("ndice de UHI")
plt.show()
```

aqui da para ver que o grfico no est legivel, logo eu irei criar mais duas colunas para horas e minutos, para podermos ter uma visualizao mais clara do indice de UHI relacionado com o horario

```python
new_df['datetime'] = pd.to_datetime(new_df['datetime'])
new_df['minute'] = new_df['datetime'].dt.minute

print(new_df) # para visualizar o DataFrame com a nova coluna 'hour'
```

```python
sns.scatterplot(x="minute", y="UHI Index", data=new_df)

plt.title("Grfico de disperso entre data e indice de UHI")
plt.xlabel("Data e Hora")
plt.ylabel("ndice de UHI")
plt.show()
```

Nesse grafico podemos contestar que a maior variao do indice UHI ocorre a partir das 15:30, at as 15:40, isso engolobando varias regies
